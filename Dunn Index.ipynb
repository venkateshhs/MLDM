```python
#=---------Function to find max distance within a cluster 
import numpy as np
#euclidean Distance
dunnin_max = []
def dunnindex_max(x,y):
    e = 0.0
    dunnin_sep_max = []
    for i in range(len(x)-1):
        for j in range(i+1,len(x)):
            e += pow((float(x[i])-float(x[j])),2)  #euclidean distance
            e += pow((float(y[i])-float(y[j])),2)
            e = np.sqrt(e)
            dunnin_max.append(e)
            dunnin_sep_max.append(e)
            print("(",x[i],",",y[i],")&(",x[j],",",y[j],"):",e)
            e = 0.0
    #print("dunnin seperate:",dunnin_sep_max)
    print("-----------------------------------------------")
    
#=---------Function to find Min within all the cluster  
dunnin_min = []
def dunnindex_min(x1,y1,x2,y2):
    dist = 0.0
    dunnin_sep_min = []
    for i in range(len(x1)):
        for j in range(len(x2)):
            dist += pow((float(x1[i])-float(x2[j])),2)  #euclidean distance
            dist += pow((float(y1[i])-float(y2[j])),2)
            dist = np.sqrt(dist)
            print("(",x1[i],",",y1[i],")&(",x2[j],",",y2[j],"):",dist)
            dunnin_min.append(dist)
            dunnin_sep_min.append(dist)
            dist = 0.0
    #print("dunnin seperate min:",dunnin_sep_min)    
    print("-----------------------------------------------")
```
